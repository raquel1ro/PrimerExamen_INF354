```python
import pandas as pd
import numpy as np
#lectura de datos por categoria
datos=pd.read_csv("heart_attack_dataset.csv")
datos.head(100)
#datos.sample(n=10)
```

```python
from sklearn.impute import SimpleImputer
import numpy as np
datos_num=datos[['Age','Cholesterol','Heart Rate','Diabetes','Family History','Smoking','Obesity','Alcohol Consumption', 'Previous Heart Problems', 'Medication Use', 'Stress Level']]
array1=np.array(datos_num)

# array1 = np.asarray([[1,2,np.nan], [np.nan, 2,3], [4,6,8]])
imputacion = SimpleImputer(missing_values = np.nan, strategy = 'mean')
array2 = imputacion.fit_transform(array1)
print(array1)
print(array2)
" -- Escalado"

print('-- mostrar Escalado')
from sklearn.preprocessing import StandardScaler
escalado = StandardScaler()
array3 = escalado.fit_transform(array2)
print(array3)


" -- Normalizado"
print('-- mostrar Normalizado')
from sklearn.preprocessing import Normalizer
preprocesar = Normalizer()
array4 = preprocesar.fit_transform(array2)
print(array4)
```
