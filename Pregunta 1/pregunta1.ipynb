```python
import pandas as pd
import numpy as np
#lectura de datos por categoria
datos=pd.read_csv("heart_attack_dataset.csv")
datos.head(100)
#datos.sample(n=10)
```

**Realice el cálculo de la media, moda, cuartiles de datos y de percentiles por columna; explique qué significa en cada caso graficando los resultados (sin numpy y pandas).**

```python
import numpy as np
import math
datos_num=datos[['Age','Cholesterol','Heart Rate','Diabetes','Family History','Smoking','Obesity','Alcohol Consumption', 'Previous Heart Problems', 'Medication Use', 'Stress Level']]
matriz=np.array(datos_num)
n=len(matriz)
medias=[]
medianas=[]
modas=[]
quartiles=[]
percentiles=[]
for c in range(11):
  columna=[fila[c] for fila in matriz]
  #media
  medias.append(sum(columna)/n)
  #mediana
  columna.sort()
  if(n%2==0):
    medianas.append((columna[n//2]+columna[n//2 - 1])/2)
  else:
    medianas.append(columna[(n-1)//2])
  #Moda
  modaCol=[]
  frecModa=0
  frec=0;
  for i in range(n):
    if(i+1==n or columna[i]!=columna[i+1]):
      frec+=1
      if(frec>frecModa):
        frecModa=frec
        modaCol.clear()
        modaCol.append(columna[i])
      else:
        if(frec==frecModa):
          modaCol.append(columna[i])
      frec=0
    else:
      frec+=1
  modas.append(modaCol)
  #Quartiles
  q=[]
  for i in range(3):
    x=(i+1)/4*(n+1)
    parte_entera=int(x)
    parte_decimal=abs(x)-abs(int(x))
    if(parte_decimal>0):
      q1=columna[parte_entera-1]+(parte_decimal*(columna[parte_entera]-columna[parte_entera-1]))
    else:
      q1=columna[parte_entera-1]
    q.append(q1)
  quartiles.append(q)
  #Percentiles
  p85=math.ceil((85*n)/100)
  percentiles.append(columna[p85-1])
print("Y COMO RESULTADOS TENEMOS:")
print("Mediana indica el valor promedio de mediciones tomadas en cuanto a los diagnosticos de ataque al coraz'on, en otros atributos de cada paciente")
print("medias: ",medias,"\n")
print("medianas indica el valor que se ubica al centro de mediciones tomadas en cuanto a los diagnosticos de ataque al coraz'on, en otros atributos de cada paciente")
print("medianas: ",medianas,"\n")
print("modas indica el valor mas frecuente de mediciones tomadas en cuanto a los diagnosticos de ataque al coraz'on, en otros atributos de cada paciente")
print("modas: ",modas,"\n")
print("quartiles dividen a la distribución en cuatro partes iguales indicado el valor de mediciones en esos puntos ")
print("quartiles: ",quartiles,"\n")
print("quartiles cuando dividen a la distribución en 100 partes indicado el valor de mediciones en esos puntos ")
print("percentiles: ",percentiles,"\n")
```
